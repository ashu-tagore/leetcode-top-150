# 27. Remove Element

Given an integer array **nums** and an integer **val**, remove all occurrences of **val** in **nums** [in-place](https://en.wikipedia.org/wiki/In-place_algorithm). The order of the elements may be changed. Then return **the number of elements in nums which are not equal to val**.

Consider the number of elements in **nums** which are not equal to **val** be **k**, to get accepted, you need to do the following things:

1. Change the array **nums** such that the first **k** elements of **nums** contain the elements which are not equal to **val**. The remaining elements of **nums** are not important as well as the size of **nums**.
2. Return **k**.

If all assertions pass, then your solution will be **accepted**.

## Example 1

**Input:** `nums = [3,2,2,3]`, `val = 3`  
**Output:** `2`, `nums = [2,2,_,_]`  
**Explanation:** Your function should return `k = 2`, with the first two elements of `nums` being `2`. It does not matter what you leave beyond the returned `k` (hence they are underscores).

## Example 2

**Input:** `nums = [0,1,2,2,3,0,4,2]`, `val = 2`  
**Output:** `5`, `nums = [0,1,4,0,3,_,_,_|]`  
**Explanation:** Your function should return `k = 5`, with the first five elements of `nums` containing `0, 0, 1, 3, and 4`. Note that the five elements can be returned in any order. It does not matter what you leave beyond the returned `k` (hence they are underscores).

## Constraints

- `0 <= nums.length <= 100`
- `0 <= nums[i] <= 50`
- `0 <= val <= 100`

````python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k
````

## Intuition
I need to remove all occurrences of a target value from an array **in-place** and return the count of remaining elements. The simplest approach is to use two pointers - one to read through the array and another to write only the elements we want to keep at the front. This way, we don't need extra space or complex swapping logic.

### Approach
- Initialize a pointer `k = 0` that represents both:
  - The count of elements not equal to `val` found so far
  - The next available position to write a kept element

- Loop through the array with index `i` from 0 to `len(nums)-1`:
  - If `nums[i]` is **NOT** equal to `val`, it's an element we want to keep
  - Copy this element to position `k` (overwriting whatever was there)
  - Increment `k` by 1 to point to the next write position
  - If `nums[i]` **IS** equal to `val`, we simply skip it (don't write it to the front)

- After processing all elements, return `k` which now contains:
  - The total count of elements not equal to `val`
  - The length of the valid prefix in the array

### Complexity
**Time complexity:**  
O(n) where n is the length of nums. We traverse the array exactly once.

**Space complexity:**  
O(1) as we only use a single integer variable `k` for pointer/counting, regardless of input size.

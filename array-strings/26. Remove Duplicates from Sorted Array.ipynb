# Remove Duplicates from Sorted Array

Given an integer array `nums` sorted in **non-decreasing order**, remove the duplicates **in-place** such that each unique element appears only once. The relative order of the elements should be kept the same.

Consider the number of unique elements in `nums` to be `k`. After removing duplicates, return the number of unique elements `k`.

> The first `k` elements of `nums` should contain the unique numbers in sorted order. The remaining elements beyond index `k - 1` can be ignored.

## Constraints

- `1 <= nums.length <= 3 * 10^4`
- `-100 <= nums[i] <= 100`
- `nums` is sorted in **non-decreasing order**

---

>  **Note:** Since the array is sorted, duplicates will always be adjacent — this enables an efficient two-pointer approach with **O(n)** time and **O(1)** space complexity.

````python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        w = 1
        for r in range(1, len(nums)):
            if nums[r] != nums[r - 1]:
                nums[w] = nums[r]
                w += 1
        return w
````

## Intuition
The array is already sorted, so all duplicates are adjacent. I don't need to search the whole array for duplicates — I just need to detect when a new unique value appears and write it to the front. The simplest way is two pointers: a slow writer that only moves when a new unique value is found, and a fast reader that scans every element.

### Approach
- Initialize a write pointer `w = 1` that represents both:
  - The count of unique elements found so far
  - The next available position to write a new unique element
  - Index `0` is always unique by default, so `w` starts at `1`

- Loop through the array with index `r` from `1` to `len(nums)-1`:
  - If `nums[r]` is **NOT** equal to `nums[r-1]`, a new unique element is found
  - Copy this element to position `nums[w]` (overwriting whatever was there)
  - Increment `w` by 1 to point to the next write position
  - If `nums[r]` **IS** equal to `nums[r-1]`, it's a duplicate — skip it

- After processing all elements, return `w` which now contains:
  - The total count of unique elements
  - The length of the valid deduplicated prefix in the array

### Complexity
**Time complexity:**  
O(n) where n is the length of nums. We traverse the array exactly once with the read pointer `r`.

**Space complexity:**  
O(1) as we only use two integer variables `w` and `r` as pointers, regardless of input size.
